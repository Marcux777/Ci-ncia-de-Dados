# Deviradas Simbolicas - SymPy

Derivação Simbólica: Envolve o cálculo exato da derivada utilizando manipulação de expressões matemáticas. É realizada por meio de álgebra simbólica, onde as funções são tratadas como expressões simbólicas e as derivadas são calculadas analiticamente, resultando em uma fórmula exata para a derivada

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

def function(x):
  return sp.sin(x)

def function_np(x):
  return np.sin(x)
```

```python

x = sp.symbols('x')
f = function(x)

der_f = sp.diff(f, x)

d = sp.lambdify(x, der_f, 'numpy')
g = sp.lambdify(x, f, 'numpy')

x_vals = np.linspace(-10, 10, 10000)
y_vals = d(x_vals)
y_gvals = g(x_vals)

plt.figure(figsize=(12, 8))
plt.plot(x_vals, y_vals, label='Derivada', linestyle = '--')
plt.plot(x_vals, y_gvals, label = 'Original',linestyle = '--')
plt.legend()
plt.xlabel('x')
plt.ylabel('f\'(x)')
plt.title('Gráfico da Derivada')
plt.grid(True)
plt.show()
```

# Derivadas númericas - Numpy

Derivação Numérica: Envolve a aproximação da derivada por meio de diferenças finitas. É usada quando não é possível ou prático calcular a derivada de forma simbólica. A derivação numérica estima o valor da derivada em um ponto específico, utilizando valores conhecidos da função em pontos próximos, e é particularmente útil para dados experimentais ou quando a função é definida numericamente

```python
# Definir a função e o intervalo de x
x = np.linspace(-10, 10, 1000)
y = function_np(x)

# Calcular a derivada numérica de y
dy = np.gradient(y, x)

# Plotar a função original e sua derivada
plt.figure(figsize=(12, 8))
plt.plot(x, dy, label='Derivada', linestyle = '--')
plt.plot(x, y, label='Original', linestyle = '--')
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da Função e sua Derivada')
plt.grid(True)
plt.show()
```
