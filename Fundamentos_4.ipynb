```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp


# Define a função
def f(x):
    return x**4 - 12*x**3 + 47*x**2 - 60*x

x = np.linspace(-20, 20, 400)

y = f(x)

plt.figure(figsize=(19, 8))
plt.plot(x, y, label='$f(x) = x^4 - 12x^3 + 47x^2 - 60x$')

plt.title('Gráfico da função $f(x)$')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')

plt.ylim(-30, 50)


plt.legend()

plt.grid(True)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.show()
```

```python
x_sym = sp.symbols('x')

def df(x_val):
    der_f = sp.diff(f(x_sym), x_sym)  # Calcula a derivada simbólica de f
    d_num = sp.lambdify(x_sym, der_f, 'numpy')  # Converte para uma função numérica
    return d_num(x_val)

def newton_raphson(f_sym, df, x0, tol):
    f_num = sp.lambdify(x_sym, f_sym, 'numpy')  # Converte f para uma função numérica
    derivada = df(x0)
    cont = 0
    while abs(f_num(x0)) > tol:
        cont+=1
        x0 = x0 - f_num(x0)/derivada
    return [x0, cont]
```

Para $x_{0} = $ 0.5

```python
lista = newton_raphson(f(x_sym), df, 0.5, 1e-5)
print(f"A raiz é {lista[0]} e o número de iterações é: {lista[1]}")
```

Para $x_{0} = $ 1

```python
lista = newton_raphson(f(x_sym), df, 1, 1e-5)
print(f"A raiz é {lista[0]} e o número de iterações é: {lista[1]}")
```

Para $x_{0} = $ 2

```python
lista = newton_raphson(f(x_sym), df, 2, 1e-5)
print(f"A raiz é {lista[0]} e o número de iterações é: {lista[1]}")
```

Para $x_{0} = $ 3.4556

```python
lista = newton_raphson(f(x_sym), df, 3.4556, 1e-5)
print(f"A raiz é {lista[0]} e o número de iterações é: {lista[1]}")
```
